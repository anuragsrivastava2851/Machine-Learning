```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
# we can also use other classifiers...
```

```python
data=pd.read_excel("iris_train.xlsx")
data.head(7)
```

```python
#here sepallength,sepalwidth,petallength,petalwidth are features 

# and flower are labels 
```

```python
x=data.iloc[:,:-1]
y=data.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
model=RandomForestClassifier()
```

```python
#model.fit(x,y)
model.fit(x.values,y.values)

# x.values and y.values are used to convert them into numpy array....
```

```python
predictions=model.predict([[6.5,3,5.5,1.8]])
```

```python
predictions
```

```python
dataset=pd.read_excel("iris_test.xlsx")
```

```python
x_test=dataset.iloc[:,:-1]
```

```python
x_test.head(5)
```

```python
predictions=model.predict(x_test.values)
```

```python
predictions
```

```python
y_test=dataset.iloc[:,-1]
```

# finding accuracy...

```python
y_test.values
```

```python
#  to find the accuracy of the prediction first we find the y_test( which are the actual lables of the test dataset) and compare with the predictions...
```

```python
actual = y_test.values
```

```python
count=0
for i in range(0, len(actual)):
    if actual[i]==predictions[i]:
        count=count+1

print(f"{(count*100)/len(actual)} %")        
        
```

### Accuracy = Correct Predictions / Total Predictions
