```python
import pandas as pd 
import numpy as np
```

```python
df=pd.read_csv("housing.csv")
```

```python
df.head()
```

```python
df["income_cat"] = pd.cut(df["median_income"],bins=[0, 1.5, 3.0, 4.5, 6.0, np.inf],labels=[1, 2, 3, 4, 5])
```

```python
df.head()
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
```

```python
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
 # n_splits shows that how many splits we want...
```

```python
split.split(df, df["income_cat"])
```

```python
for train_index, test_index in split.split(df, df["income_cat"]):
    strat_train_set = df.loc[train_index]
    strat_test_set = df.loc[test_index]
```

```python
strat_train_set.head()
```

```python
df=strat_train_set.copy()
```

## Impute missing value

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy="median")
```

```python
housing_num=housing_features.select_dtypes(include=[np.number])  # here we select only numerical columns {we remove 'ocean_proximity' columns}
```

```python
imputer.fit(housing_num)
```

```python
imputer.statistics_
```

```python
X = imputer.transform(housing_num)
```

```python
X
```

# for categorical columns

```python
housing=pd.DataFrame(X,columns=housing_num.columns,index=housing_num.index)
```

```python
housing['ocean_proximity']=df['ocean_proximity']
```

```python
housing
```

```python
housing=housing[['ocean_proximity']]
```

#### ordinal encoding

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
ordinal_encoder = OrdinalEncoder()
```

```python
housing_cat=ordinal_encoder.fit_transform(housing)
```

```python
housing_cat=pd.DataFrame(housing_cat,columns=housing.columns,index=housing.index)
```

```python
housing_cat
```

```python
# ⚠️ Caution: Ordinal encoding implies an order between categories, 
# which may not be true here. For example, 
# it treats INLAND (1) as closer to <1H OCEAN (0) than NEAR OCEAN (4), which might not make sense.
```
